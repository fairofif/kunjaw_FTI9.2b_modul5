```python
from plotly import __version__
import cufflinks as cf
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
cf.set_config_file(sharing='public',theme='pearl',offline=True)
```

bebas pake library apapun buat di kelas aku, tapi output valuenya sama ya :) disini aku pake plotly biar diagramnya responsif kalo cursornya diarahin ke diagramnya bakal ada popup detail valuenya. Secara logika sama aja, cuma beda nama fungsinya dan variable2nya.

case 1

```python
print('coba arahin cursornya ke bar chartnya, nanti keluar detailsnya')
df3[df3["Attrition_Flag"] == "Existing Customer"]["Total_Revolving_Bal"].iplot(kind="hist", yTitle = 'Frequency', xTitle = 'Nominal',color='#68e8cc',bins=20)
```

case 2

```python
dfedulevel = df3[df3["Total_Trans_Ct"] > 30]["Education_Level"].value_counts().rename_axis('Education_Level').reset_index(name='counts')
print("Coba arahin cursor nya ke pie chartnya, nanti keluar detailnya")
print("kalian juga bisa klik kategori sebelah kanan buat ngeilangin beberapa kategori")

fig = px.pie(values = dfedulevel["counts"], names = dfedulevel["Education_Level"])
fig.show()
```

case 3

```python
df3[df3["Gender"] == "M"].iplot(kind="scatter", mode="markers", x = "Customer_Age", y = "Credit_Limit", color='#68e8cc')
```

case 4

```python
dfcategorycard = df3["Card_Category"].value_counts().rename_axis('Card_Category').reset_index(name='counts')
dfcategorycard.iplot(kind="barh", y = "counts", x = "Card_Category")
```

case 5

```python
dfmaritalstatus = df3[df3["Marital_Status"].isin(["Married","Single","Divorced"])]

dfplot = dfmaritalstatus.groupby(["Gender","Marital_Status"]).size().reset_index()
dfplot.rename(columns = {0:'counts'}, inplace=True)

fig = px.bar(dfplot, x="Gender",y="counts", color="Marital_Status")
fig.show()
```
